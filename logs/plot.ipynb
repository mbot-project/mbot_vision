```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV files
df_memryx = pd.read_csv('metrics_memryx_bank.csv')
df_pi = pd.read_csv('metrics_pi_bank.csv')

# Add system identifier
df_memryx['system'] = 'Memryx'
df_pi['system'] = 'Pi'

# Display the first few rows of each dataframe
print("Memryx Data:")
print(df_memryx.head())
print("\nPi Data:")
print(df_pi.head())
```

```python
# Create FPS over time plot
plt.figure(figsize=(12, 6))
plt.plot(df_memryx['elapsed_time_sec'], df_memryx['fps'], label='Memryx', linewidth=2)
plt.plot(df_pi['elapsed_time_sec'], df_pi['fps'], label='Pi', linewidth=2)
plt.title('FPS over Time', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Frames Per Second', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create CPU utilization over time plot
plt.figure(figsize=(12, 6))
plt.plot(df_memryx['elapsed_time_sec'], df_memryx['cpu_percent'], label='Memryx', linewidth=2)
plt.plot(df_pi['elapsed_time_sec'], df_pi['cpu_percent'], label='Pi', linewidth=2)
plt.title('CPU Utilization over Time', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('CPU Utilization (%)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create memory usage plot
plt.figure(figsize=(12, 6))
plt.plot(df_memryx['elapsed_time_sec'], df_memryx['memory_used_mb'], label='Memryx', linewidth=2)
plt.plot(df_pi['elapsed_time_sec'], df_pi['memory_used_mb'], label='Pi', linewidth=2)
plt.title('Memory Usage over Time', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Memory Used (MB)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create memory percentage plot
plt.figure(figsize=(12, 6))
plt.plot(df_memryx['elapsed_time_sec'], df_memryx['memory_percent'], label='Memryx', linewidth=2)
plt.plot(df_pi['elapsed_time_sec'], df_pi['memory_percent'], label='Pi', linewidth=2)
plt.title('Memory Usage Percentage over Time', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Memory Usage (%)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
